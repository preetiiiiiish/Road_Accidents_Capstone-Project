**4.	Offenders and victims who died according to gender as well the as the total deaths.**

```python

columns_for_pie_chart = [
    'Offenders (Driver/Pedestrian) Died_Male',
    'Offenders (Driver/Pedestrian) Died_Female',
    'Offenders (Driver/Pedestrian) Died_Transgender'
]
total_sum = df1['Offenders (Driver/Pedestrian) Died_Total'].sum()

percentages = (df1[columns_for_pie_chart] / total_sum) * 100

labels = ['Died Male', 'Died Female', 'Died Transgender']

values = percentages.sum()

colors = ['#1f78b4', '#FF8C00', 'lightgreen']
explode=[0,0.1,0]

plt.figure(figsize=(5,5))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=colors, startangle=360,explode=explode,shadow=True)
plt.title('Distribution of Offenders Died by Gender')
plt.show()
```

```python
columns_for_pie_chart = [
    'Victims Died_Male',
    'Victims Died_Female',
    'Victims Died_Transgender'
]
total_sum = df1['Victims Died_Total'].sum()
percentages = (df1[columns_for_pie_chart] / total_sum) * 100
labels = ['Victim_Died_Male', 'Victim_Died_Female', 'Victim_Died_Transgender']
values = percentages.sum()

colors = ['#1f78b4', '#FF8C00', 'lightgreen']
explode=[0,0.1,0]

plt.figure(figsize=(5,5))
plt.pie(values, labels=labels, autopct='%1.4f%%', colors=colors, startangle=360,explode=explode,shadow=True)
plt.title('Percentage of victim according to Gender who died')
plt.show()
```

```python
columns_for_pie_chart = ['Offenders (Driver/Pedestrian) Died_Total', 'Victims Died_Total']
total_sum = df1[columns_for_pie_chart].sum().sum()
percentages = (df1[columns_for_pie_chart].sum() / total_sum) * 100
labels = ['Offenders (Driver/Pedestrian) Died_Total', 'Victims Died_Total']
values = percentages.values
colors = ['#FF8C00', '#1f78b4']
explode=[0,0.1]
plt.figure(figsize=(4,4))
plt.pie(values, labels=labels, autopct='%1.2f%%', colors=colors, startangle=135,explode=explode,shadow=True)
plt.title('Percentage of total offenders and victims who died in accidents')
plt.show()
```
